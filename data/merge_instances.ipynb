Will YOLO work better when we merge back-to-back GT instances together?

Only one way to find out...

```python
from pathlib import Path
import pandas as pd
import shutil
```

```python
csv_fns = list(Path('./LeitmotifOccurrencesInstances/Instances').rglob('*.csv'))
len(csv_fns)
```

```python
merge_threshold = 0.3
```

```python
for fn in csv_fns:
    merged_dir = fn.parent.parent.parent / 'MergedInstances' / fn.parent.name
    merged_dir.mkdir(exist_ok=True, parents=True)
    new_data = {
        "Motif": [],
        "StartSec": [],
        "EndSec": []
    }
    df = pd.read_csv(fn, sep=';')

    last_motif = ''
    last_start = 0
    last_end = 0
    merging = False
    for i, row in df.iterrows():
        motif, start, end = row['Motif'], float(row['StartSec']), float(row['EndSec'])
        if motif == last_motif and start - last_end < merge_threshold:
            last_end = end
            merging = True
        else:
            if merging:
                new_data['Motif'].append(last_motif)
                new_data['StartSec'].append(last_start)
                new_data['EndSec'].append(last_end)
                merging = False
            else:
                new_data['Motif'].append(motif)
                new_data['StartSec'].append(start)
                new_data['EndSec'].append(end)
            last_motif = motif
            last_start = start
            last_end = end

    if merging:
        new_data['Motif'].append(last_motif)
        new_data['StartSec'].append(last_start)
        new_data['EndSec'].append(last_end)

    new_df = pd.DataFrame(new_data)
    print(f"{fn.parent.name}/{fn.name} - Original: {len(df.index)} - Merged: {len(new_df.index)}")
    new_df.to_csv(merged_dir / fn.name, sep=';', index=False)

```
